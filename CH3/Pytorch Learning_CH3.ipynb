### CH3. Logistic Regression

#### 1. Sigmoid function (feat.Matplotlib)

```python
%matplotlib inline
# 브라우저 내부(inline)에 바로 그려주도록 명령
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x,y,'g')
# Green Sigmoid function
plt.plot([0,0.0], [0,1.0], ':')
# Blue dot line at x = 0
plt.title("Sigmoid function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
## Since we declare %matplotlib inline, we actually do not need to use plt.show() function.
```
